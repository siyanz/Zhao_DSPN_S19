1. How does knowing data type (e.g., qualitative vs. quantitative) provide insights into the type of distribution a random variable comes from?

**Answer:** 

In most cases, if the data type is quantitative, we assume that the data is a good representation of the population. A random variable from such a sample is most likely (though not necessarily the case) to be normally distributed. If the data type is qualitative (e.g., treatment group), a random variable that is from a qualitative dataset is not assumed to be normally distributed.

**Answer:** 

It means that the data point comes from a pool of values that is drawn from a Gaussian distribution. The mean of the pool of values is 100 and the deviation around the mean is 15. It also mean that it is highly likely that the value of the data point is between 70 - 130.

3. Using the *runif* function, run three experiments by simulating the outcomes of rolling a single, six-sided die. Show the distribution of each experiment. Show how the simulated means compare to the expected mean of a fair roll.
        	a) Exp 1: 10 independent throws
        	b) Exp 2: 1,000 independent throws
        	c) Exp 3: 10,000 independent throws


```r
# -------------------------------
# Question 3
# -------------------------------
# set the size of the plot in jupyter
options(repr.plot.width=3, repr.plot.height=2) 
library("ggplot2")#load the library


# Exp 1
tenThrows <- data.frame(val = floor(runif(10, min = 1, max = 7)))
tenMean <- mean(tenThrows$val)
ggplot(tenThrows, aes(x=val)) +
    geom_histogram(fill='white',color='black', bins=6) + labs(title = "Exp 1") #histogram


# Exp 2
thouThrows <- data.frame(val = floor(runif(1000, min = 1, max = 7)))
thouMean <- mean(thouThrows$val)
ggplot(thouThrows, aes(x=val)) +
    geom_histogram(fill='white',color='black', bins=6) + labs(title = "Exp 2") #histogram

# Exp 3
tenThouThrows <- data.frame(val = floor(runif(10000, min = 1, max = 7)))
tenThouMean <- mean(tenThouThrows$val)
ggplot(tenThouThrows, aes(x=val)) +
    geom_histogram(fill='white',color='black', bins=6) + labs(title = "Exp 3") #histogram


sprintf("Exp 1: %.1f", tenMean)
sprintf("Exp 2: %.1f", thouMean)
sprintf("Exp 3: %.1f", tenThouMean)

sprintf("As the number of throws increases, the similated mean of the roll approaches 3.5 (which is the expected mean).")
```
